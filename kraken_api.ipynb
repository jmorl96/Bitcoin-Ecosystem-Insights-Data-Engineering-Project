```python
import requests

pair = "XBTUSD"
since = "1735171200"

url = f"https://api.kraken.com/0/public/Trades"

payload = {}
query_params = {"pair": pair, "since": since}
headers = {
  'Accept': 'application/json'
}

response = requests.request("GET", url, headers=headers, data=payload, params=query_params)

print(response.text)
```

```python
response.raise_for_status()
```

```python
response.json()["result"].keys()
```

```python
for key in response.json()["result"].keys():
    if key != "last":
        data = response.json()["result"][key]
#last = response.json()["result"]["last"]
```

```python
import pandas as pd
```

```python
columns = [
    "price",
    "volume",
    "time",
    "buy_sell",
    "market_limit",
    "miscellanious",
    "trade_id"
]
```

```python
dataframe = pd.DataFrame(data=data, columns=columns)
```
